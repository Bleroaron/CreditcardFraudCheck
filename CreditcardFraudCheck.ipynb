## Part A

In the rapidly evolving digital age, credit card fraud has emerged as a significant threat,  largely fueled by the rapid expansion of digital payment systems and e-commerce. These can be difficult for a human to detect. Implementing machine learning techniques to reduce the substantial financial losses individuals and institutions alike suffer is crucial for enhancing the security and integrity of financial transactions. The goal of this review is to provide a comprehensive analysis and review of current strategies in fraud detection, their success rates, and potential areas for improvement, thereby contributing to the ongoing efforts in safeguarding financial transactions against fraudulent exploits.

Credit cards are safeguarded through measures like data encryption and tokenization, yet these aren't foolproof against fraud. When it comes to detecting fraudulent activities at scale, the task becomes considerably complex. A significant volume of transaction data is required to effectively train and implement fraud detection systems. Moreover, this data is often skewed; the overwhelming majority of transactions are legitimate, which makes the rare instances of fraud harder to identify.This skewness necessitates the use of techniques like the Synthetic Minority Oversampling Technique (SMOTE), a method for balancing data by generating synthetic samples. SMOTE helps in distinguishing rare fraudulent activities amid a majority of valid transactions by creating additional, synthetic examples of the minority class.

One of the earliest works in applying machine learning for credit card fraud detection is the research conducted by Ghosh and Reilly in 1994 in collaboration with Mellon Bank. They utilised a neural network-based approach for detecting fraudulent transactions. The neural network used in the study was trained on a large dataset of credit card transactions, including both fraudulent and non-fraudulent activities. The results showed that the neural network-based system significantly outperformed traditional rule-based fraud detection methods. savings in the range of 20% to in excess of 40% can be achieved.

The neural network was trained on a large set of labelled credit card transactions, including various types of fraud such as lost cards, stolen cards, application fraud, counterfeit fraud and mail-order fraud. A unique aspect of the study was the use of a holdout data set, which consisted of all account activity over a two-month period, to test the trained model. That was preserved for the testing of the model.

Overall, the study by Ghosh and Reilly represents a significant advancement in credit card fraud detection, showcasing the power of data mining in helping to identify complex fraud patterns and substantially improving the accuracy and efficiency of fraud management systems.

In their 2022 study, Ileberi, Sun, and Wang proposed a machine learning-based approach for credit card fraud detection, emphasising the use of the Genetic Algorithm (GA) for feature selection. This method addresses a critical aspect of machine learning - the identification of the most relevant features from large datasets to improve the efficiency and accuracy of predictive models. In the context of credit card transactions, where vast amounts of data are generated, selecting key features that contribute to fraud detection is most important. The GA, inspired by evolution, iteratively mutates features to identify the most effective set for fraud detection. This process is crucial as it not only enhances the model's performance but also reduces computational complexity by eliminating redundant or irrelevant data.

The results of this study are particularly significant. The study demonstrated that the model outperformed existing systems in detecting credit card fraud. The model was validated using a dataset generated from European cardholders, suggesting its effectiveness in the real-world. This outcome emphasises the potential of integrating traditional machine learning techniques with evolutionary algorithms like GA for feature optimization.

Both the ANN and RF algorithms achieved a high test accuracy of 99.94%. With relevant features: V1, V5, V7, V8, V11,V13, V14, V15, V16, V17, V18, V19, V20, V21, V22, V23, V24, Amount

The RF method showed the best results in terms of precision and accuracy, achieving a fraud detection accuracy of 99.98% and a precision of 95.34%. With features: Time, V1, V7, V8, V9, V11, V12, V14, V15, V22, V27, V28, Amount

These statistics clearly demonstrate the effectiveness of the RF method, particularly when combined with GA for feature selection, in accurately detecting credit card fraud.

### 1. Artificial Neural Networks (ANN)

Strengths and Advantages:

ANNs have several strengths and advantages that make them particularly effective in dynamic environments like credit card transaction processing. One of their key strengths is adaptive learning, as ANNs have the capability to learn and improve from new data continuously. This feature is crucial in environments where data patterns and behaviours are constantly evolving. Additionally, ANNs excel in pattern recognition; they are exceptionally good at identifying complex patterns and relationships between different features. This ability is vital for detecting fraud that might not be apparent through simpler analytical methods. Another significant advantage of ANNs is their scalability. They are well-suited to handle the vast volumes of data typical in credit card transactions, making it a well suited choice for large-scale fraud detection applications.

Weaknesses and Disadvantages:

ANNs, while powerful, come with certain limitations. One of the primary challenges is their complexity; they require significant computational resources and a high level of expertise for effective setup and training. This can make them less accessible for smaller organisations or those with limited technical capabilities. Additionally, ANNs are often criticised for their opaque decision-making processes. They are frequently referred to as "black boxes" because it can be challenging to understand and interpret the internal workings and how they arrive at specific decisions. Finally, the performance of ANNs is heavily dependent on the quantity and quality of the training data. They require large, clean datasets to function optimally, and their effectiveness can decrease if the data is sparse, noisy, or biassed.

Application:

ANNs are applied in credit card fraud detection to  analyse various aspects of transaction data, such as the amount, location, and frequency of transactions. Their primary function is to detect unusual patterns that could signify fraudulent activities. For effective training, ANNs require an extensive dataset with a wide range of past transactions, including both fraudulent and legitimate transactions. This extensive dataset allows the network to learn and identify differences between normal and suspicious transaction behaviours.

Input data required/expected output:

Regarding data requirements and output, ANNs need large amounts of transactional data, which must be accurately labelled to denote fraudulent or legitimate transactions. This labelling is important as it guides the learning process of the network, enabling it to distinguish between different types of transactions effectively. The expected output from an ANN in this context is a probability for each transaction. This score represents the likelihood that a given transaction is fraudulent. Based on this probabilistic assessment, transactions can be flagged for further action.

### 2. Random Forest (RF)

Strengths and Advantages:

RF algorithm is known for its high accuracy in classification problems, crucial for correctly identifying fraudulent transactions. Additionally, RF is good at handling overfitting, which is a common issue with complex models, by building various trees on different subsets of data and then averaging their predictions. RF also excels in identifying the importance of datasets and features, ranking the attributes used in classification, such as transaction amount, location, and frequency. This not only aids in making accurate predictions but also provides valuable insights into the factors most indicative of fraudulent activity, which is not only essential for understanding and improving fraud detection techniques but help reduce storage needs.

Weaknesses and Disadvantages:

RF models, due to the nature of constructing multiple decision trees, can become quite large and complex, leading to significant computational demands, especially when processing large-scale data. While they are simpler than models like ANNs, RF models are still not as straightforward as algorithms such as logistic regression, making it challenging to trace the exact decision-making path. Furthermore, with very large datasets typical in credit card transaction processing; the training time for RF models tends to increase with the size of the dataset, as each tree in the forest requires training on a subset of the data. This can result in slower training times, thus making it less suitable where rapid deployment or frequent updates are necessary.

Application:

RF algorithm is adept at analysing a variety of transaction data, such as transaction amount, time, location, and user behaviour, to effectively identify potential fraud. One of its key strengths lies in its adaptability to ever-changing fraud patterns, making it particularly effective in dynamic environments where fraudulent tactics frequently evolve. RF's ability to quickly adapt and respond to new types of fraud is a significant advantage over simpler, less flexible models.

Input data required/expected output:

RF relies on structured data where features are extracted from detailed transaction information. Input data does not need to be normalised as it wouldnâ€™t lead to a different result. The expected output from an RF model is an average taken from the output of the different trees created from the datasets. This binary classification enables quick and decisive action to be taken on transactions flagged as potentially fraudulent.

### 3. K-Nearest Neighbors (KNN)

Strengths and Advantages:

KNN stands out for its simplicity, as it is a straightforward and easily implementable algorithm. Additionally, KNN is highly flexible, proving effective in classification problems and showing adaptability to any changes in the input data. This adaptability is particularly beneficial in dynamic environments where data patterns can shift rapidly. A characteristic of KNN is the absence of a model training phase, a feature not common in other algorithms. This means that KNN operates directly on raw data.

Weaknesses and Disadvantages:

As the dataset size increases, the scalability of KNN becomes a concern. KNN tends to slow down significantly with larger datasets. This limitation is primarily due to its need to compute distances for each new data point relative to all existing points, a process that becomes increasingly time-consuming as the dataset grows which can be a significant hurdle in scenarios where fast, real-time predictions are essential. Furthermore, KNN's performance is sensitive to irrelevant or redundant features in the data. If the data includes such features, the algorithm's ability to accurately classify new points can diminish, leading to poorer performance. This necessitates careful feature selection and preprocessing to ensure the relevance and quality of the data being used.

Application:

KNN's method of comparing a given transaction with a set of known transactions makes it particularly suitable for scenarios where fraud patterns are in a state of constant evolution. Unlike many other algorithms that rely on a fixed model structure, KNN's flexibility lies in its ability to adapt to new data without the need for retraining a model. By not being constrained to a predefined model structure, KNN can more readily adjust to patterns, making it a valuable technique to prevent credit card fraud.

Input data required/expected output:

For the KNN algorithm to function effectively in fraud detection, the input data typically consists of transactional data characterised by various features. These features, such as transaction amount, location, and time, are essential as they are used to calculate the distances between different transactions. The distance calculations enable the algorithm to identify and compare the similarities between transactions. The expected output from KNN comes from the closest transactions to the one being analysed and classifies it based on the most common classification among these neighbours. This method uses the idea that similar transactions are likely to have the same class label, thus providing a basis for the algorithm to make informed classifications.

```python
#method to iterrate through NNs and calculate scores
def trainModel():
    models = {}
    for index in range(7, 8):
        model = modelSelection(index)
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        # Train the model
        trained_model = model.fit(x_smote, y_smote, epochs=10, batch_size=256, validation_data=(x_test, y_test))

        # Evaluate the model
        y_pred = model.predict(x_test)
        y_pred = (y_pred > 0.5)

        # Calculate performance metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        auc = roc_auc_score(y_test, y_pred)

        # Storing model and its data
        models[index] = {'data': {"accuracy": accuracy, "precision": precision, "recall": recall, "f1": f1, "auc": auc},
                         'model': trained_model,'predict':y_pred}

        # Print performance metrics
        print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1-Score: {f1}, AUC: {auc}")

    return models

result = trainModel()
```

Accuracy (99.97): This metric indicates the proportion of the total number of predictions that were correct. An accuracy of approximately 99.97% means that almost all predictions made by the model were correct. However, accuracy alone can be misleading, especially if the dataset is imbalanced.

Precision (82.32): Precision measures the ratio of true positive predictions to the total number of positive predictions (true positives plus false positives). A precision of about 82.42% means that when the model predicts an instance to be positive, there is an 82.42% chance it is correct. A lower precision indicates a higher number of false positives.

Recall (99.26): Also known as sensitivity, it measures the ratio of true positive predictions to the total actual positives (true positives plus false negatives). A recall of 99.96% means the model correctly identified almost all actual positive instances in the test dataset.

F1 Score (90): The F1 score is a balanced metric that considers both false positives and false negatives. An F1 score of approximately 90.37% is quite high and suggests that the model has a good balance between precision and recall. This is particularly important in scenarios where both false positives and false negatives are crucial, such as fraud detection.

AUC (Area Under the Curve) (99.61): AUC refers to the area under the ROC (Receiver Operating Characteristic) curve. It measures the ability of the model to distinguish between classes. An AUC of approximately 99.98% indicates excellent model performance, with a high capability to differentiate between positive and negative instances.


```python
def plot(data):
  plt.figure(figsize=(12, 6))

  # Plot accuracy
  plt.subplot(1, 2, 1)
  plt.title('Model Accuracy')
  plt.xlabel('Epoch')
  plt.ylabel('Accuracy')
  plt.plot(data.history['accuracy'])
  plt.plot(data.history['val_accuracy'])
  plt.legend(['Train', 'Validation'], loc='upper left')

  # Plot loss
  plt.subplot(1, 2, 2)
  plt.title('Model Loss')
  plt.xlabel('Epoch')
  plt.ylabel('Loss')
  plt.plot(data.history['loss'])
  plt.plot(data.history['val_loss'])
  plt.legend(['Train', 'Validation'], loc='upper left')
  plt.tight_layout()
  plt.show()
```
